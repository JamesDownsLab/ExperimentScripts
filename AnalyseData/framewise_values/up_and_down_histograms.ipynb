```python
from Generic import filedialogs
from ParticleTracking import dataframes
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = filedialogs.load_filename()
data = dataframes.DataStore(file)
```

```python
duty = data.df.groupby(data.df.index).first().Duty.values
```

```python
%matplotlib auto
plt.plot(duty)
plt.xlabel('frame')
plt.ylabel('Duty / 1000')
```

```python
frame_means = data.df.groupby(data.df.index).mean()
duty = frame_means.Duty
order = frame_means.order_mag
```

```python
fig, ax = plt.subplots()
ax.plot(duty, order)
```

```python
duty_order = frame_means[['order_mag', 'Duty']]
duty_order
```

```python
duty_max = duty_order.Duty.max()
duty_order = duty_order.reset_index()
```

```python
duty_order.head()
```

```python
frames = duty_order.loc[duty_order.Duty==duty_max, 'frame'].values
frames_switch = round(frames.mean())
frames_switch
```

```python
duty_order_up = duty_order.loc[duty_order.frame < frames_switch, ['Duty', 'order_mag', 'frame']]
duty_order_down = duty_order.loc[duty_order.frame >= frames_switch, ['Duty', 'order_mag', 'frame']]
```

```python
order_up = duty_order_up.groupby('Duty')['order_mag'].mean()
order_down = duty_order_down.groupby('Duty')['order_mag'].mean()
order_up.head()
```

```python
fig, ax = plt.subplots()
ax.plot(order_up.index, order_up.values, label='up')
ax.plot(order_down.index, order_down.values, label='down')
ax.legend()
ax.set_xlabel('Duty Cycle / 1000')
ax.set_ylabel('Order parameter')
ax.set_title(file)
```

```python
all_up = data.df.loc[data.df.index < frames_switch]
all_up.head()
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return array[idx]
```

```python
all_up = all_up.reset_index().set_index('Duty')
all_up
```

```python
duties = [400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 950, 1000]
fig, ax = plt.subplots()
for di in duties:
    d = find_nearest(duty_order_up.index, di)
    orders = all_up.loc[d, 'order_mag'].values
    freq, bins = np.histogram(orders, bins=100)
    ax.plot(bins[:-1], freq, label=d)
ax.legend(title='Duty Cycle')
ax.set_xlabel('Order')
ax.set_ylabel('Frequency')
ax.set_title(file)
```
